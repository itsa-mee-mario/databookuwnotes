```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# define data
x = 3 # true slope
a = np.arange(-2, 2, 0.25)

a = a.reshape(-1, 1)


b = x*a + np.random.randn(*a.shape)
```

```python
# plot the line and the data
plt.plot(a, b, 'o')
plt.plot(a, x*a, '-')
plt.show()

```

```python
U, S, VT = np.linalg.svd(a, full_matrices=False)
xtilde = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ b

```

```python
# plot a line with slope xtilde
plt.plot(a, b, 'o')
plt.plot(a, x*a, '--', color='blue') # true value
plt.plot(a, xtilde*a, '-', color='red') # estimated value
plt.show()
```
