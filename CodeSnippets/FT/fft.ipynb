```python
# create a sin wave
import numpy as np
x = np.arange(0, 3*np.pi, 0.05)
f = np.sin(x)
```

```python
import matplotlib.pyplot as plt
plt.plot(x, f)
```

```python
def DFT(f):
    N = len(f)
    F = np.zeros(N, dtype=complex)
    for k in range(N):
        for n in range(N):
            F[k] += f[n]*np.exp(-2j*np.pi*k*n/N)
    return F
```

```python
def close_enough(a, b, tol=1e-3):
    return abs(a-b) < tol
```

```python
close_enough(DFT(f), np.fft.fft(f))
```

```python
# extract component frequencies and magnitudes
def get_freqs(f, fs):
    N = len(f)
    freqs = np.arange(0, N)*fs/N
    magnitudes = np.abs(f)
    return freqs, magnitudes
```

```python
plt.hist(get_freqs(DFT(f), 100))
```

```python
# plot the spectrum
plt.plot(get_freqs(DFT(f), 100)[0], get_freqs(DFT(f), 100)[1])
```
