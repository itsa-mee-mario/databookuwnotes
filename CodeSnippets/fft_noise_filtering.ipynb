```python
# noise filtering with FFT
import numpy as np
```

```python
t = np.linspace(0, 10, 1000)
f = np.sin(2 * np.pi * t) + np.sin(3 * np.pi * t) + np.random.randn(len(t))
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15, 5)
plt.plot(t, f)
```

```python
# get the FFT of the signal, and plot the magnitude
F = np.fft.fft(f)
plt.plot(np.abs(F[:len(t) // 2]))
```

```python
# remove the smaller frequencies (with magn less than 1)
T = 200 # tolarence
F[np.abs(F) < T] = 0
```

```python
plt.plot(np.abs(F[:len(t) // 2]))
```

```python
# inverse FFT
f_filtered = np.fft.ifft(F)
plt.plot(t, f_filtered)
```
