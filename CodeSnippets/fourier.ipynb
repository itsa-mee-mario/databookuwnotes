```python
import numpy as np
```

```python

dx = 0.001
L = np.pi
x = np.arange(-L, L, dx)
```

```python
n = len(x)
nquarter = int(n/4)

# define hat function 
f = np.zeros(n)
f[nquarter:nquarter*2] = x[2*nquarter:3*nquarter]
f[nquarter*2:nquarter*3] = x[nquarter*3:2*nquarter:-1]
# plot f
import matplotlib.pyplot as plt
plt.plot(x, f)
```

```python
# try to to approximate f using fourier series
# f = sum(a_n * cos(n * x) + b_n * sin(n * x)) for n = 0, 1, 2, ...

a0 = f*np.ones_like(x)*dx
fFs = a0/2
for k in range(20):
    a = f@np.cos(np.pi*k*x/L) * dx
    b = f@np.sin(np.pi*k*x/L) * dx
    fFs += a*np.cos(np.pi*k*x/L) + b*np.sin(np.pi*k*x/L)
    # create an animation of the fourier series
    plt.plot(x, fFs)
   
    


```
