```python
# read the csv file wine.data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

wine_data = pd.read_csv('wine.data', header=None)
```

```python
wine_data_without_labels = wine_data.iloc[:, 1:]
```

```python
wine_data_without_labels_matrix = np.array(wine_data_without_labels)
```

```python
# apply normalization
wine_data_without_labels_matrix_normalized = (wine_data_without_labels_matrix - np.mean(wine_data_without_labels_matrix, axis=0)) / np.std(wine_data_without_labels_matrix, axis=0)
```

```python
scaler = StandardScaler()
df_wine_norm = pd.DataFrame(scaler.fit_transform(wine_data_without_labels_matrix_normalized))
df_wine_norm.head()
```

```python
# add back the labels to the dataframe
df_wine_norm['class'] = wine_data.iloc[:, 0]
```

```python
df_wine_norm
```

```python
df_wine_norm.describe()
```

```python
U, S, V = np.linalg.svd(df_wine_norm.values[:, :13], full_matrices=False)
```

```python
# plot the singular values
plt.figure()
plt.plot(range(1, len(S) + 1), S)
plt.xlabel('Singular Value')
plt.ylabel('Value')
plt.show()
```

```python
# plot cumulative sum of the singular values
plt.figure()
plt.plot(range(1, len(S) + 1), np.cumsum(S) / np.sum(S))
plt.xlabel('Singular Value')
plt.ylabel('Cumulative Sum')
# plot a line when the cumulative sum is 90%
plt.axhline(y=0.90, color='r', linestyle='--')
plt.show()
```

```python
U.shape, S.shape, V.shape
```

```python
# transform the data into 2D space by mat mul the data with V.T
df_wine_norm_2d = pd.DataFrame(np.dot(df_wine_norm.values[:, :13], V.T))
```

```python

plt.figure()
plt.scatter(df_wine_norm_2d.iloc[:, 0], df_wine_norm_2d.iloc[:, 1], c=df_wine_norm.values[:, 13], cmap='viridis')
plt.show()
```

```python
# for the first 5 singular values,
# for every pair of the singular values,
# plot the data. 
# display as a grid of subplots
plt.figure( figsize=(20, 20))

for i in range(5):
    for j in range(i+1):
        plt.subplot(5, 5, i*5 + j + 1)
        plt.scatter(df_wine_norm_2d.iloc[:, i], df_wine_norm_2d.iloc[:, j], c=df_wine_norm.values[:, 13], cmap='viridis')
        plt.xlabel('x-axis')
        plt.ylabel('y-axis')
        plt.title('SV' + str(i*5 + j + 1))
```
