```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
A = imread('a.jpg')
```

```python
plt.imshow(A)
```

```python
X = np.mean(A, -1) # convert rgb to greyscale by taking the average of the three channels
```

```python
U, S, V = np.linalg.svd(X, full_matrices=False)
S = np.diag(S)
img = plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
```

```python
j = 0
for r in (5, 20, 50, 100):
    Xapprox = U[:,:r] @ S[:r,:r] @ V[:r,:]
    plt.figure(j+1)
    j+=1
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.axis('off')
    plt.show()
```

```python
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title('Singular values')
plt.show()


plt.figure(2)
plt.plot(np.cumsum(np.diag(S))/np.sum(np.diag(S)))
plt.title('Cumulative sum of singular values')
plt.show()
```
